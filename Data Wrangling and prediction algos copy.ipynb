## Deriving the deltas

Process data for each fund, adding quarter and year. We would want to do delta analysis on ticker's values, percentage, and shares. 

```python
def get_funds():
    funds = []
    for i in range(1, 9):
        df = pd.read_csv('data/funds/fund'+str(i)+'.csv', parse_dates=['date'])
        df = df.dropna(subset=['date', 'type'])
        df = pd.merge(df, snp[['from', 'thru', 'co_tic']], left_on='ticker', right_on='co_tic', how='outer')
        if 'change_type' in Set(df.columns):
            df['change_type'] = df['change_type'].fillna('no_change')
        df['snp'] = df.apply(lambda x: 1 if not pd.isnull(x['from']) and not pd.isnull(x['thru']) and not pd.isnull(x['date']) and x['date'] >= x['from'] and x['date'] <= x['thru'] else 0, axis=1)
        df['year'] = df['date'].apply(lambda x: x.year)
        df['quarter'] = df['date'].apply(lambda x: (x.month-1)//3 + 1)
        df['change_shares'] = df.apply(lambda x: x['current_shares'] - x['previous_shares'], axis=1)
        df['change_value'] = df.apply(lambda x: x['current_value'] - x['previous_value'], axis=1)
        if 'previous_percent' in Set(df.columns):
            df['change_percent'] = df.apply(lambda x: x['current_percent'] - x['previous_percent'], axis=1)
            df['change_percent'] = df['change_percent'].fillna(0)
        df['change_shares'] = df['change_shares'].fillna(0)
        df['change_value'] = df['change_value'].fillna(0)
        funds.append(df)
    return funds
funds = get_funds()
```

## Linear Regression

But is there really no correlation between price change and change in holdings??? 
We can test it out with linear regression on changes in ticker prices -- is the strategy closely tied with market prices?

First define a linear_regression wrapper that can be reused. 

```python
colors = ['r','g','b']
def linear_regression_wrapper(df, indep_list, dep):
    X_HD=df[indep_list + [dep]]
    X_HD = X_HD.fillna(0)
    X_HDn = X_HD.apply(lambda x: (x - x.mean())/x.std(), axis=0)
    dep_std=X_HDn[dep].values.reshape(-1,1)
    indep_std = [X_HDn[ind].values.reshape(-1,1) for ind in indep_list]
    indep_std_transpose = np.array(indep_std).transpose()
    X_train, X_test, y_train, y_test = train_test_split(indep_std_transpose[0], dep_std, test_size=0.1, random_state=42)
    clf1 = LinearRegression()
    clf1.fit(X_train, y_train)
    predicted_train = clf1.predict(X_train)
    predicted_test = clf1.predict(X_test)
    
    color_ind = 0
    fig = plt.figure()
    plt1 = fig.add_subplot(2, 1, 1)
    plt1.set_title (str(indep_list)+ "&" + delt + "_train")
    for ind in xrange(X_train.shape[1]):
        plt1.scatter(X_train[:,ind], predicted_train, c=colors[ind])
        color_ind = color_ind +1    
    
    color_ind = 0
    plt2 = fig.add_subplot(2, 1, 2)
    plt2.set_title (str(indep_list)+ "&" + delt + "_test")
    fig.tight_layout()
    for ind in xrange(X_test.shape[1]):
        plt2.scatter(X_test[:,ind], predicted_test, c=colors[ind])
        color_ind = color_ind +1      
    print "These are the coefficients and intersection: ", (clf1.coef_, clf1.intercept_)
    print "Train scores on the prediction: ", clf1.score(X_train, y_train)
    print "Test scores on the prediction: ", clf1.score(X_test, y_test)
    return (clf1.coef_, clf1.intercept_)
```

## Bayesian prediction

If we really want to be able to predict, we can run Bayesian prediction on the percentage holdings for currently held tickers on quarterly partitioned data. 

```python
npoll = len(fund.groupby(['year', 'quarter']).groups)
fund['current_percent'] = fund['current_percent'].fillna(0)
ticker_groups = fund.groupby(['ticker'])
ticker_summary= ticker_groups.mean()
ticker_summary['mean'] = ticker_summary['current_percent']
tau = np.std(ticker_summary['mean'])
ticker_summary['se'] = np.std(ticker_summary['mean'])/np.sqrt(npoll)
B = (1/tau**2)/(1/ticker_summary['se']**2+1/tau**2)
B_df = pd.DataFrame(B, columns=['B'])
B_df.reset_index(inplace=True)
ticker_summary['mu_post'] = (1-B)*ticker_summary['mean']
ticker_summary.reset_index(inplace=True)
plt.scatter(ticker_summary['mean'], ticker_summary['mu_post'], s=50)
plt.title("Bayes vs. Raw  Averages")
plt.xlabel("Raw Average")
plt.ylabel("Bayes")
currentx = plt.xlim()
currenty = plt.ylim()
plt.plot((-1,1),(-1,1), c='black', linewidth=1)
plt.xlim(currentx)
plt.ylim(currenty)
plt.show()
```

```python
ticker_summary['sd_post'] = np.sqrt((1-B_df['B'])*ticker_summary['se']**2)
```

With calculated posterior mu and sd, we can run Monte Carlo simulation 100 times and take the average on percentage for each ticker for the tickers that are in the fund's portfolio in the 3rd quarter of 2014. Then we can predict what the percentage holding for each ticker is for the next quarter. When we get data from the next quarter, we can validate whether our prediction is close.

```python
third_quarter = ticker_summary[(ticker_summary["year"] == 2014) & (ticker_summary["quarter"] == 3)].sort('mean', ascending=False)
third_quarter.reset_index(inplace=True)
NSIM = 100
simarr = np.empty((len(third_quarter.index), ))

for i in xrange(NSIM):
    simulated = np.random.normal(third_quarter['mu_post'], scale=third_quarter['sd_post'], size =len(third_quarter.index))
    simulated[simulated < 0] = 0
    simarr = np.vstack((simarr, np.array(simulated)))

third_quarter['prediction'] = pd.Series(simarr.mean(axis=0))
third_quarter[['ticker','prediction']].head()
```
