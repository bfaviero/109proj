```python
%matplotlib inline
import matplotlib.pyplot as plt
from __future__ import division
import itertools
import numpy as np
import pandas as pd # pandas
import datetime
from pyq_api import get_ticker_info
import os
from utils import *
from sqlalchemy import Column, ForeignKey, String, Date, Float, Integer
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship
from sqlalchemy import create_engine
from process_data import get_funds
from utils import get_quarters
from sqlalchemy.orm import sessionmaker
from sqlalchemy import UniqueConstraint
from numpy import nan
```

**Do some processing to get unique funds, tickers, and sectors.**

```python

funds = get_funds()[:7]
soros = get_funds()[7]
funds = [merge_with_quarters(fund) for fund in funds]
soros = merge_with_quarters(soros)
quarters_df = get_quarters()
def merge_with_quarters(df):
    return df.merge(quarters_df, left_on='date', right_index=True)
unique_tickers = set()
for fund in funds:
    for ticker in fund.ticker.values:
        unique_tickers.add(ticker)
fund_ticker_sector = dict()
for fund in funds[:7]:
    for ticker, sector in fund[['ticker', 'sector']].values:
        if ticker not in fund_ticker_sector:
            fund_ticker_sector[ticker] = set([sector])
        else:
            fund_ticker_sector[ticker].add(sector)
def get_sector(row):
    ticker = row['ticker']
    if ticker in fund_ticker_sector:
        s = fund_ticker_sector[ticker]
        return s.pop() if s else 'None'
    return 'None'
soros['sector'] = soros.apply(lambda x: get_sector(x), axis=1)
allfunds = funds + [soros]
merged = pd.concat(allfunds)
merged = merged.fillna(-1)
merged.sector = merged.sector.replace(' ', 'None')
merged.sector = merged.sector.replace(-1, 'None')
merged = merged.loc[merged.sector != 'None']
sectors = merged.sector.unique()
filers = merged.filer.unique()
```

**The merged dataframe contains all of the fund data needed.**

###This section will explore the relationship between the funds and the different sectors they invest in.

**Here we calculate the percent of a fund that each sector takes up in a given quarter. That is, as seen in the head below, Information Technology was 32% of Coatue's fund value in Q1 of 2001.**

```python
df = merged.loc[(merged.filer == filer)].groupby('date').sum()[[]]

final_df = pd.DataFrame()
value_dfs = []
for filer in filers:
    for sector in sectors:
        df2 = merged.loc[(merged.filer == filer) & (merged.sector == sector)]
        value_df = pd.DataFrame(merged.loc[(merged.filer == filer) & (merged.sector == sector)].groupby('date').current_value.sum(), columns=['portfolio_sector_value'])
        value_df['total_portfolio_value'] = pd.DataFrame(merged.loc[merged.filer == filer].groupby('date').current_value.sum(), columns=['total_portfolio_holdings'])
        value_df['percent'] = value_df.portfolio_sector_value * 1.0 / value_df.total_portfolio_value
        value_df['sector'] = sector
        value_df['filer'] = filer
        value_df = value_df.fillna(0)[['percent', 'sector', 'filer']]
        value_dfs.append(value_df)
percents = pd.concat(value_dfs).merge(quarters_df, left_index=True, right_index=True).merge
percents.head()
```

**We then plot a matrix of graphs below showing how the allocation of the funds' investments by sector reacts to the performance of certain sectors.**


```python
def plot_percent_of_fund_matrix():
    def clean_labels(labels):
        labels = [str(label) for label in labels]
        used = set()
        for i in range(len(labels)):
            label = labels[i]
            if label in used:
                labels[i] = ""
            else:
                used.add(label)
        return labels
    f = plt.figure(figsize=(22, 20))
    columns = 8
    rows = 14
    for column, filer in enumerate(filers):
        for row, sector in enumerate(sectors):
            num_plot = (column + 1) + row * 8
            sp = f.add_subplot(rows, columns, num_plot)
            if num_plot <=8:
                sp.set_title(filer[:17], y=1)
            if column==0:
                sp.set_ylabel(sector, rotation=0)
            else:
                sp.set_ylabel("")
            sp.set_ylim([0, 1])
            sp.set_yticks([])
            df = percents.loc[(percents.sector == sector) & (percents.filer == filer)][['percent', 'year']]
            df['data'] = 1
            sector_value = pd.DataFrame(merged.loc[(merged.sector == sector)].groupby('date').current_value.sum(), columns=['portfolio_sector_value'])
            subdf = sector_value.merge(df, left_index=True, right_index=True, how='left').fillna(0)
            subdf['year'] = [d.year for d in subdf.index.values]
            subdf.portfolio_sector_value = subdf.portfolio_sector_value * subdf.data
            labels = clean_labels(subdf.year.values)
            num_labels = len(labels)

            sp.set_xticks(range(num_labels))
            sp.set_xticklabels([""]*num_labels)
            sp.plot(subdf.percent, color='black')
            value_axis = sp.twinx()
            value_axis.set_xticks([])
            value_axis.set_yticks([])
            value_axis.plot(subdf.portfolio_sector_value, color='red')
            value_axis.legend(loc='center left', bbox_to_anchor=(1.1, 0.9))

            correlation = subdf.loc[subdf.data==1].corr().iloc[0][1]
            average_percent = subdf.loc[subdf.data == 1].percent.mean()
            if np.isnan(average_percent) or average_percent > 100:
                average_percent = np.nan
            else:
                average_percent = int(average_percent * 1000)/10.0

            if np.isnan(correlation):
              correlation = np.nan 
            hedge_fund_sector_data.append((filer, sector, correlation, average_percent))

            #plot 2008
            try:
                index1 = labels.index('2008')
                index2 = index1 + 1
                while not labels[index2]:
                    index2+=1
                value_axis.vlines(x=[index1, index2], ymin=0, ymax=subdf.portfolio_sector_value.max(), linestyle='dotted')
            except:
                pass
```

```python
def plot_correlation_boxplots():
    sector_metrics = pd.DataFrame(hedge_fund_sector_data, columns = ['firm', 'sector', 'correlation', 'average_percent_invested'])
    sector_metrics.correlation = sector_metrics.correlation.replace('1', np.nan)
    sector_metrics.correlation = sector_metrics.correlation.replace(1, np.nan)

    sector_metrics.average_percent_invested = sector_metrics.average_percent_invested.replace(-1, np.nan)
    sector_metrics.average_percent_invested = sector_metrics.average_percent_invested.replace('-1', np.nan)

    sector_metrics_no_na = sector_metrics.dropna()

    #make correlation absolute value, since a fund might always bet against a sector
    sector_metrics_no_na.correlation.values
    sector_metrics_no_na.correlation = sector_metrics_no_na.correlation.abs()
    f = plt.figure(figsize=(20, 2))
    sector_metrics_no_na.boxplot(column='correlation', by='firm', rot=90) 
```

```python
def plot_average_percent_plot():
    f = plt.figure(figsize=(20, 2))
    sector_metrics_no_na.boxplot(column='average_percent_invested', by='firm', rot=90)
```

**In the graph below:**
* the red line is the sector's performance over time
* The black line signifies what percent of that fund's value is taken up by that sector
* The dotted lines represent 2008, so we can see if there are any marked differences before, during, and after the financial crisis.

##Analysis
From [This paper](http://depts.washington.edu/sce2003/Papers/284.pdf) lays down a few criteria for hedge fund. In the graphs below, we see two types of hedge funds:


* **Market neutral**: aggressively use leverage to capitalize on small pricing differences. Attempt to minimize market risk by staying sector-neutral and actively seek out opportunities for profit in any sector.
* **Market Tactical**: Place longer-term bets on sectors to and capitalize on early mispricing.  

We see that coatue maintains a constantly high percentage of its fund in IT, and its percentage of IT and consumer discretionary are inversely correlated. That is, it has most of its money either in IT or consumer discretionary. It is interesting to note, since Coatue considers itself primarily a technology firm, but we see that it has decreased its investment in IT steadily since 2008. 

Cayman, Elliott, and Tudor maintain relatively stable proportions of certain sectors, making them more passive, longer-term investors, with a few exceptions such as Tudor decreasing their Finance percentage after the financial crisis but rapidly building that position back up around a year later.

Wellington tracks sectors more closely, as evidenced by its higher correlation, but also spreads those investments more evenly, trying to distribute risk, roughly maintaining the distribution of sectors in its portfolio.

Soros is known as one of the most successful hedge funds that exists. Soros seems to track sectors less closely, and based on the data from the matrix as well throws its leverage aggressively behind any market that seems like it's doing well. 
Soros seems to build up no long-term bets. The "average percent" boxplot supports this and the matrix as well, as Soros, when the fund does invest, seems to use its much of its leverage and take a higher risk in that specific position by doing so.
