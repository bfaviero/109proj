```python

funds = get_funds()[:7]
soros = get_funds()[7]
```

```python
funds = [merge_with_quarters(fund) for fund in funds]
soros = merge_with_quarters(soros)
```

```python
quarters_df = get_quarters()
def merge_with_quarters(df):
    return df.merge(quarters_df, left_on='date', right_index=True)
```

```python
unique_tickers = set()
for fund in funds:
    for ticker in fund.ticker.values:
        unique_tickers.add(ticker)
```

```python
fund_ticker_sector = dict()
for fund in funds[:7]:
    for ticker, sector in fund[['ticker', 'sector']].values:
        if ticker not in fund_ticker_sector:
            fund_ticker_sector[ticker] = set([sector])
        else:
            fund_ticker_sector[ticker].add(sector)
```

```python
def get_sector(row):
    ticker = row['ticker']
    if ticker in fund_ticker_sector:
        s = fund_ticker_sector[ticker]
        return s.pop() if s else 'None'
    return 'None'
```

```python
soros['sector'] = soros.apply(lambda x: get_sector(x), axis=1)
```

```python
allfunds = funds + [soros]
merged = pd.concat(allfunds)
```

```python
def clean_labels(labels):
    labels = [str(label) for label in labels]
    used = set()
    for i in range(len(labels)):
        label = labels[i]
        if label in used:
            labels[i] = ""
        else:
            used.add(label)
    return labels
```

```python
f = plt.figure(figsize=(22, 20))
columns = 8
rows = 14
for column, filer in enumerate(filers):
    for row, sector in enumerate(sectors):
        num_plot = (column + 1) + row * 8
        sp = f.add_subplot(rows, columns, num_plot)
        if num_plot <=8:
            sp.set_title(filer[:17], y=1)
        if column==0:
            sp.set_ylabel(sector, rotation=0)
        else:
            sp.set_ylabel("")
        sp.set_ylim([0, 1])
        sp.set_yticks([])
        df = percents.loc[(percents.sector == sector) & (percents.filer == filer)][['percent', 'year']]
        df['data'] = 1
        sector_value = pd.DataFrame(merged.loc[(merged.sector == sector)].groupby('date').current_value.sum(), columns=['portfolio_sector_value'])
        subdf = sector_value.merge(df, left_index=True, right_index=True, how='left').fillna(0)
        subdf['year'] = [d.year for d in subdf.index.values]
        subdf.portfolio_sector_value = subdf.portfolio_sector_value * subdf.data
        labels = clean_labels(subdf.year.values)
        num_labels = len(labels)
        
        sp.set_xticks(range(num_labels))
        sp.set_xticklabels([""]*num_labels)
        sp.plot(subdf.percent, color='black')
        value_axis = sp.twinx()
        value_axis.set_xticks([])
        value_axis.set_yticks([])
        value_axis.plot(subdf.portfolio_sector_value, color='red')
        value_axis.legend(loc='center left', bbox_to_anchor=(1.1, 0.9))
        
        correlation = subdf.loc[subdf.data==1].corr().iloc[0][1]
        average_percent = subdf.loc[subdf.data == 1].percent.mean()
        if np.isnan(average_percent) or average_percent > 100:
            average_percent = 'x'
        else:
            average_percent = int(average_percent * 1000)/10.0
            
        if np.isnan(correlation):
          correlation = 'x'  
        sp.text(-12, 0.8, str(correlation)[:4], fontsize=12)
        sp.text(-12, 0.66, str(average_percent), fontsize=12)
        #plot 2008
        try:
            index1 = labels.index('2008')
            index2 = index1 + 1
            while not labels[index2]:
                index2+=1
            value_axis.vlines(x=[index1, index2], ymin=0, ymax=subdf.portfolio_sector_value.max(), linestyle='dotted')
        except:
            pass
        
```
