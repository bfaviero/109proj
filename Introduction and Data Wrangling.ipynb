```python
def get_tickerdb_data(ticker):
    session = establish_session()
    data = [(e.date, e.price) for e in session.query(Ticker).filter(Ticker.name==ticker).all()]
    if data:
        df = pd.DataFrame(data=data, columns=['date', 'price'])
        df['change_price'] = df['price'].diff()
        df['year'] = df['date'].apply(lambda x: x.year)
        df['quarter'] = df['date'].apply(lambda x: (x.month-1)//3 + 1)
        return df
    else:
        return None
```

```python
## process data for each fund
def get_funds():
    funds = []
    for i in range(1, 9):
        df = pd.read_csv('data/funds/fund'+str(i)+'.csv', parse_dates=['date'])
        df = df.dropna(subset=['date', 'type'])
        df = pd.merge(df, snp[['from', 'thru', 'co_tic']], left_on='ticker', right_on='co_tic', how='outer')
        if 'change_type' in Set(df.columns):
            df['change_type'] = df['change_type'].fillna('no_change')
        df['snp'] = df.apply(lambda x: 1 if not pd.isnull(x['from']) and not pd.isnull(x['thru']) and not pd.isnull(x['date']) and x['date'] >= x['from'] and x['date'] <= x['thru'] else 0, axis=1)
        df['year'] = df['date'].apply(lambda x: x.year)
        df['quarter'] = df['date'].apply(lambda x: (x.month-1)//3 + 1)
        df['change_shares'] = df.apply(lambda x: x['current_shares'] - x['previous_shares'], axis=1)
        df['change_value'] = df.apply(lambda x: x['current_value'] - x['previous_value'], axis=1)
        if 'previous_percent' in Set(df.columns):
            df['change_percent'] = df.apply(lambda x: x['current_percent'] - x['previous_percent'], axis=1)
            df['change_percent'] = df['change_percent'].fillna(0)
        df['change_shares'] = df['change_shares'].fillna(0)
        df['change_value'] = df['change_value'].fillna(0)
        funds.append(df)
    return funds
funds = get_funds()
```

```python
#first test things out with ticker prices -- is the strategy closely tied with market prices?
## indep_list of indices/prices that can affect our model
colors = ['r','g','b']
def linear_regression_wrapper(df, indep_list, dep):
    X_HD=df[indep_list + [dep]]
    X_HD = X_HD.fillna(0)
    X_HDn = X_HD.apply(lambda x: (x - x.mean())/x.std(), axis=0)
    dep_std=X_HDn[dep].values.reshape(-1,1)
    indep_std = [X_HDn[ind].values.reshape(-1,1) for ind in indep_list]
    indep_std_transpose = np.array(indep_std).transpose()
    X_train, X_test, y_train, y_test = train_test_split(indep_std_transpose[0], dep_std)
    clf1 = LinearRegression()
    clf1.fit(X_train, y_train)
    predicted_train = clf1.predict(X_train)
    predicted_test = clf1.predict(X_test)
    color_ind = 0
    fig_train = plt.figure()
    fig_train.suptitle(str(indep_list)+ "&" + delt + "_train", fontsize=20)
    for ind in xrange(X_train.shape[1]):
        plt.scatter(X_train[:,ind], predicted_train, c=colors[ind], alpha=0.5)
        color_ind = color_ind +1    
    
    color_ind = 0
    fig_test = plt.figure()
    fig_test.suptitle(str(indep_list)+ "&" + delt + "_test", fontsize=20)
    for ind in xrange(X_test.shape[1]):
        plt.scatter(X_test[:,ind], predicted_test, c=colors[ind], alpha=0.5)
        color_ind = color_ind +1      
    print "These are the coefficients and intersection: ", (clf1.coef_, clf1.intercept_)
    print "Train scores on the prediction: ", clf1.score(X_train, y_train)
    print "Test scores on the prediction: ", clf1.score(X_test, y_test)
    return (clf1.coef_, clf1.intercept_)

linear_regression_wrapper(smaller_frame, ['cpi', 'gdp', 'inflation'], 'change_shares')
```
