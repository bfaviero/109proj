```python
## process data for each fund
def get_funds():
    funds = []
    for i in range(1, 9):
        df = pd.read_csv('data/funds/fund'+str(i)+'.csv', parse_dates=['date'])
        df = df.dropna(subset=['date', 'type'])
        df = pd.merge(df, snp[['from', 'thru', 'co_tic']], left_on='ticker', right_on='co_tic', how='outer')
        df = pd.merge(df, sectors, on='ticker', how='inner')
        if 'change_type' in Set(df.columns):
            df['change_type'] = df['change_type'].fillna('no_change')
        df['snp'] = df.apply(lambda x: 1 if not pd.isnull(x['from']) and not pd.isnull(x['thru']) and not pd.isnull(x['date']) and x['date'] >= x['from'] and x['date'] <= x['thru'] else 0, axis=1)
        df['year'] = df['date'].apply(lambda x: x.year)
        df['quarter'] = df['date'].apply(lambda x: (x.month-1)//3 + 1)
        df['change_shares'] = df.apply(lambda x: x['current_shares'] - x['previous_shares'], axis=1)
        df['change_value'] = df.apply(lambda x: x['current_value'] - x['previous_value'], axis=1)
        funds.append(df)
    return funds

funds = get_funds()
```

```python
def row_with_return(row):
    #Add the return of that ticker starting this quarter until the beginning of next quarter
    row = row.copy()
    def apply_return(r):
        ticker = r['ticker']
        typ = r['type']
        ticker_price = r['ticker_price']
        current_shares = r['current_shares']
        current_value = r['current_value']
        next_ticker_price = r['next_ticker_price']
        returns = 0
        if next_ticker_price:
            if typ == 'SH':
                returns = (next_ticker_price - ticker_price) * current_shares
            elif typ == 'CALL':
                returns = (next_ticker_price - ticker_price) * current_shares
            elif typ == 'PUT':
                returns = (ticker_price - next_ticker_price) * current_shares
        return returns
    row['returns'] = row.apply(lambda r: apply_return(r), axis=1)
    return row
```

```python
def row_with_next_ticker_price(row):
    #add the ticker price at the beginning of the next quarter
    row = row.copy()
    def apply_next_ticker_price(r):
        ticker = r['ticker']
        quarter = r['quarter']
        year = r['year']
        quarter += 1
        if quarter == 5:
            year += 1
            quarter = 1
        next_quarter_dates = q.loc[(q.quarter == quarter) & (q.year == year)].date.values
        if not next_quarter_dates:
            return r['ticker_price']
        next_quarter_date = next_quarter_dates[0]
        next_ticker_price = get_price_of_stock_on_or_around_day(ticker, next_quarter_date)
        return next_ticker_price
    row['next_ticker_price'] = row.apply(lambda r: apply_next_ticker_price(r), axis=1)
    return row
```

```python
def get_quarter_dfs(fund):
    #get a dataframe for the fund where each df is the data for a specific quarter
    fund_quarters = []
    for year in fund.year.unique():
        for quarter in (1, 2, 3, 4):
            if np.isnan(year):
                continue
            row = fund.loc[(fund.quarter==quarter) & (fund.year == year)]
            if row:
                fund_quarters.append(row)
    return fund_quarters
```

```python
def get_sectors_over_time(unique_sectors, fund_quarters):
    #Get a list of the number of shares in each sector for each quarter
    sector_dict = {}
    totals = []
    for sector in unique_sectors:
        sector_dict[sector] = []
    for fund_quarter in fund_quarters:
        quarter_sectors = fund_quarter.groupby('sector').count().sector.to_dict()
        total = sum(quarter_sectors.values())
        totals.append(total)
        for sector in unique_sectors:
            sector_dict[sector].append(quarter_sectors.get(sector, 0))
    if ' ' in sector_dict:
        del sector_dict[' ']
    for key in sector_dict.keys():
        if np.isnan(key):
            del sector_dict[key]
            break
    return total, sector_dict
```

```python
def get_sectors_over_time_by_value(unique_sectors, fund_quarters):
    #get the total value of each sector in a fund for a given quarter
    sector_dict = {}
    totals = []
    for sector in unique_sectors:
        sector_dict[sector] = []
    for fund_quarter in fund_quarters:
        quarter_sectors = fund_quarter.reset_index()[['current_value', 'sector']].groupby('sector').sum().to_dict()['current_value']
        total = sum(quarter_sectors.values())
        totals.append(total)
        for sector in unique_sectors:
            sector_dict[sector].append(quarter_sectors.get(sector, 0))
    if ' ' in sector_dict:
        del sector_dict[' ']
    for key in sector_dict.keys():
        if np.isnan(key):
            del sector_dict[key]
            break
    return total, sector_dict
```

```python
def graph_sectors_over_time():
    for fund in funds:
        if 'sector' in fund:
            total, sector_dict = get_sectors_over_time_by_value(fund.sector.unique(), get_quarter_dfs(fund))
            values = [np.array(value) * 1.0 / total for value in sector_dict.values()]
            for value in values:
                plt.plot(value)
            plt.show()
graph_sectors_over_time()
```

```python
def get_shares_over_time(unique_tickers, fund_quarters):
    ticker_dict = {}
    totals = []
    for ticker in unique_tickers:
        ticker_dict[ticker] = []
    for fund_quarter in fund_quarters:
        tickers = fund_quarter.set_index('ticker')[['current_shares']].to_dict()['current_shares']
        totals.append(sum(tickers.values()))
        for ticker in unique_tickers:
            ticker_dict[ticker].append(tickers.get(ticker, 0))
    print ticker_dict.keys()[:4]
    return totals, ticker_dict
```
